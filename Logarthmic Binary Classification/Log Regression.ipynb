```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("HR_comma_sep.csv")
df.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder
obj = LabelEncoder()
```

```python
df.salary = obj.fit_transform(df.salary)
df.head(20)
```

```python
df_drop = df.drop(columns=['Department'])
```

```python
dum_dep = pd.get_dummies(df['Department'])
dum_dep.head(10)
```

```python
merge = pd.concat([df_drop, dum_dep], axis=1)
merge.head(10)
```

```python
import seaborn as sns
```

```python
sns.heatmap(data=merge.head(100).corr(), annot=True, cmap = "coolwarm")
```

```python
corr_data = merge.corr()['left'].abs()
corr_data
```

```python
sns.lineplot(data=corr_data, markers='+')
```

```python
corr_data = corr_data.sort_values(ascending=False)
corr_data
```

```python
corr_data = corr_data[(corr_data > 0.06) | (corr_data == 0.001294)]
corr_data = pd.DataFrame(corr_data)
corr_data = corr_data.T
```

```python
testX = pd.concat([merge[corr_data.columns], merge['salary']], axis = 1)
testX = testX.drop(columns=['left'])
testX
```

```python
x = testX
y = merge.left
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)
```

```python
x.corr()
```

```python
exercise = pd.concat([ytrain, xtrain], axis = 1)
exercise
```

```python
from sklearn.linear_model import LogisticRegression
obj = LogisticRegression()
```

```python
obj.fit(X = x, y = y)
```

```python
obj.score(xtest, ytest)
```
