```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df.Sex = le.fit_transform(df.Sex)
df.ExerciseAngina = le.fit_transform(df.ExerciseAngina)
df.head()
```

```python
new_df = pd.get_dummies(df, columns=['ChestPainType', 'RestingECG', 'ST_Slope'])
new_df
```

```python
new_df.drop(['ChestPainType_ASY', 'RestingECG_LVH', 'ST_Slope_Down'], axis=1,  inplace=True)
new_df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x = new_df.drop(['HeartDisease'], axis = 1)
y = new_df.HeartDisease
```

```python
x_cols = x.columns
scaler.fit(x[x_cols])
x[x_cols] = scaler.transform(x[x_cols])
x.shape
```

SVM \
Logistic \
Random forest 

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
s, r, l = SVC(), RandomForestClassifier(), LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
def train(obj, x=xtrain, y=ytrain):
    obj.fit(xtrain, ytrain)
```

```python
l = [s, r, l]
for i in l:
    train(i)
```

```python
def getScore(obj, x=xtest, y=ytest):
    score = obj.score(xtest, ytest)
    return score
```

```python
score = {}
for i in l:
    score[i] = (getScore(i))
score
```
