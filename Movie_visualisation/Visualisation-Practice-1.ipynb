```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Hydra-Movie-Scrape.csv")
df.head(10)
```

```python
df = df[(df['Year'] >= 2005) & (df['Year'] <= 2009)]
df.shape
```

```python
df = df[(df['Rating'] >= 7) & (df['Rating'] <= 9)]
df.shape
```

```python
randomDf = df.sample(n = 20)
randomDf
```

```python
import numpy as np
x = np.arange(20)
plt.xlabel("Movie Name(Converted to index)")
plt.ylabel("Rating")
plt.title("Example of Visualiazation")
plt.plot(x, randomDf.Rating, label = 'Rating')
plt.legend(loc = 'best')
```

```python
plt.plot(x, randomDf.Rating, 'c', label = 'Rating')
plt.legend(loc = 'best')
plt.show()
```

```python
topDf = df.nlargest(n = 20, columns=["Rating"])
```

```python
topDf.head()
```

```python
plt.plot(x, topDf.Rating, label = 'Rating')
plt.legend(loc = 'best')
plt.show()
```
